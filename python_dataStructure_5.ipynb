## 2 ìŠ¤íƒì´ˆê¸°í™”

```python
SIZE = 3 # ìŠ¤íƒí¬ê¸°ì§€ì •
stack = [None for _ in range(SIZE)]
top = -1
```

## 3 ìŠ¤íƒí•¨ìˆ˜êµ¬í˜„

### 1ï¸âƒ£ Push

```python
#ğŸ“ ìŠ¤íƒì´ ê½‰ ì°¼ëŠ”ì§€ í™•ì¸í•˜ëŠ” í•¨ìˆ˜
def isStackFull():
    global SIZE,stack, top

    if (top >= SIZE - 1):
        return True # ìŠ¤íƒê½‰ì°¸
    else:
        return False # ìŠ¤íƒì—¬ìœ 
```

```python
#ğŸ“ ë°ì´í„°ë¥¼ ì‚½ì…í•˜ëŠ” í•¨ìˆ˜
def push(data):
    global SIZE, stack, top
    
    if (isStackFull()):
        print("ìŠ¤íƒì´ ê½‰ ì°¼ìŠµë‹ˆë‹¤.")
        return # ì¢…ë£Œ

    top += 1
    stack[top] = data
```

### 2ï¸âƒ£ Pop

```python
#ğŸ“ ìŠ¤íƒì´ ë¹„ì—ˆëŠ”ì§€ í™•ì¸í•˜ëŠ” í•¨ìˆ˜
def isStackEmpty():
    if (top == -1):
        return True
    else:
        return False
```

```python
#ğŸ“ ìŠ¤íƒì—ì„œ ë°ì´í„°ë¥¼ ì¶”ì¶œí•˜ëŠ” í•¨ìˆ˜
def pop():
    global SIZE, stack, top

    if (isStackEmpty()):
        print("ìŠ¤íƒì´ ë¹„ì—ˆìŠµë‹ˆë‹¤.")
        return None
    
    data = stack[top]
    stack[top] = None
    top -= 1
    return data
```

### 3ï¸âƒ£ Peek : top ìœ„ì¹˜ ë°ì´í„° í™•ì¸ë§Œ í•˜ê³  ìŠ¤íƒì— ê·¸ëŒ€ë¡œ ë‘ëŠ” ê²ƒ

```python
def peek():
	global SIZE, stack, top
	if (isStackEmpty()) :
		print("ìŠ¤íƒì´ ë¹„ì—ˆìŠµë‹ˆë‹¤.")
		return None
	return stack[top]
```
