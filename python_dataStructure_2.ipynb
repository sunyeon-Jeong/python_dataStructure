## 5ï¸âƒ£ ë°ì´í„°ì‚½ì…_í•¨ìˆ˜êµ¬í˜„

```python
media = []

def insert_data(position, media_name):
    
    # ìŒìˆ˜ or ë¦¬ìŠ¤íŠ¸ê¸¸ì´ë³´ë‹¤ í° ê²½ìš° ê²€ì—´
    if position < 0 or position > len(media): 
        print("ë°ì´í„° ì‚½ì… ìœ„ì¹˜ë¥¼ ë²—ì–´ë‚¨")
        return # í•¨ìˆ˜ì¢…ë£Œ

    media.append(None) # ê³µê°„ í™•ë³´
    mLen = len(media) # ë¦¬ìŠ¤íŠ¸ í˜„ì¬ê¸¸ì´

    for i in range(mLen-1, position, -1): # (ì´ìƒ, ë¯¸ë§Œ, ì¦ê°)
        # ë§ˆì§€ë§‰ìë¦¬ -> ë§ˆì§€ë§‰-1ìë¦¬ ê°’ ì”Œì›€
        media[i] = media[i-1]
        # ë§ˆì§€ë§‰-1ìë¦¬ Noneê°’
        media[i-1] = None

    # ì§€ì •ìœ„ì¹˜ -> ë¯¸ë””ì–´ì¶”ê°€
    media[position] = media_name

print(media)
insert_data(0,"ë¼ë°”")
insert_data(0,"ë½€ë¡œë¡œ")
print(media)
```

## 6ï¸âƒ£ ë°ì´í„°ì‚­ì œ_í•¨ìˆ˜êµ¬í˜„

```python
def del_data(position):
    if position < 0 or position >= len(media):
        print("ë°ì´í„° ì‚½ì… ìœ„ì¹˜ë¥¼ ë²—ì–´ë‚¨")
        return

    mLen = len(media) # ë¦¬ìŠ¤íŠ¸ í˜„ì¬ê¸¸ì´
    media[position] = None # ë°ì´í„°ì‚­ì œ

    for i in range(position+1, mLen):
        media[i-1] = media[i]
        media[i] = None

    # ë¦¬ìŠ¤íŠ¸ ë§ˆì§€ë§‰ ê³µê°„ ì‚­ì œ
    del(media[mLen-1])

del_data(0)
print(media)
```
