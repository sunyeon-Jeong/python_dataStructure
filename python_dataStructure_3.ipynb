## 9ï¸âƒ£ ë‹¨ìˆœì—°ê²°ë¦¬ìŠ¤íŠ¸ í•¨ìˆ˜êµ¬í˜„

```python
class Node() :
	def __init__ (self) :
		self.data = None
		self.link = None

def printNodes(start) :
	current = start
	if current == None :
		return
	print(current.data, end = ' ')
	while current.link != None:
		current = current.link
		print(current.data, end = ' ')
	print()

def insertNode(findData, insertData) :
	global memory, head, current, pre
	if head.data == findData : # ì²« ë²ˆì§¸ ë…¸ë“œ ì‚½ì…
		node = Node()
		node.data = insertData
		node.link = head
		head = node
		return

	current = head
	while current.link != None : # ì¤‘ê°„ ë…¸ë“œ ì‚½ì…
		pre = current
		current = current.link
		if current.data == findData :
			node = Node()
			node.data = insertData
			node.link = current
			pre.link = node
			return

	node = Node() # ë§ˆì§€ë§‰ ë…¸ë“œ ì‚½ì…
	node.data = insertData
	current.link = node

def findNode(findData) :
	global memory, head, current, pre

	current = head
	if current.data == findData :
		return current
	while current.link != None :
		current = current.link
		if current.data == findData :
			return current
	return Node()	# ë¹ˆ ë…¸ë“œ ë°˜í™˜

def deleteNode(deleteData) :
	global memory, head, current, pre

	if head.data == deleteData : # ì²« ë²ˆì§¸ ë…¸ë“œ ì‚­ì œ
		current = head
		head = head.link
		del(current)
		return

	current = head # ì²« ë²ˆì§¸  ì™¸ ë…¸ë“œ ì‚­ì œ
	while current.link != None :
		pre = current
		current = current.link
		if current.data == deleteData :
			pre.link = current.link
			del(current)
			return
```

```python
head, current, pre = None, None, None
memory = []
dataArray = ["ì„œìš¸", "ëŒ€ì „", "ë¶€ì‚°"]

node = Node() # ì²« ë²ˆì§¸ ë…¸ë“œ
node.data = dataArray[0]
head = node
memory.append(node)


for data in dataArray[1:] : # ë‘ ë²ˆì§¸ ë…¸ë“œë¶€í„°
    pre = node
    node = Node()
    node.data = data
    pre.link = node
    memory.append(node)


printNodes(head)
insertNode('ëŒ€ì „','ì²œì•ˆ')
insertNode('ì„œìš¸','íŒŒì£¼')
printNodes(head)
```
