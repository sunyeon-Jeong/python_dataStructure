## 4ï¸âƒ£ í í•¨ìˆ˜êµ¬í˜„ (ì¼ë°˜í)

```python
#ğŸ“ íê°€ ê½‰ì°¼ëŠ”ì§€ í™•ì¸í•˜ëŠ” í•¨ìˆ˜
def isQueueFull():
	global SIZE, queue, front, rear

	if (rear == SIZE-1):
		return True # ê½‰ì°¸
	else:
		return False
```

```python
#ğŸ“ ë°ì´í„°ë„£ëŠ” enQueue í•¨ìˆ˜
def enQueue(data):
	global SIZE, queue, front, rear

	if (isQueueFull()):
		print("íê°€ ê½‰ ì°¼ìŠµë‹ˆë‹¤.")
		return
        
	rear += 1
	queue[rear] = data
```

```python
#ğŸ“ íê°€ ë¹„ì—ˆëŠ”ì§€ í™•ì¸í•˜ëŠ” í•¨ìˆ˜
def isQueueEmpty() :
	global SIZE, queue, front, rear

	if (front == rear) :
		return True # ë¹„ì—ˆìŒ
	else :
		return False
```

```python
#ğŸ“ ë°ì´í„°ë¹¼ëŠ” deQueue í•¨ìˆ˜
def deQueue() :
	global SIZE, queue, front, rear

	if (isQueueEmpty()) :
		print("íê°€ ë¹„ì—ˆìŠµë‹ˆë‹¤.")
		return None

	front += 1
	data = queue[front]
	queue[front] = None
	return data
```

```python
#ğŸ“ í ë°ì´í„° ì¶œë ¥í•˜ëŠ” í•¨ìˆ˜
def checkQueue():
    for i in range(0, len(queue), 1):
        print(queue[i], end = ' ')
```

```python
#ğŸ“ peekí•¨ìˆ˜
def peek():
    global SIZE, queue, front, rear

    if (isQueueEmpty()):
        print("íê°€ ë¹„ì—ˆìŠµë‹ˆë‹¤.")
        return None

    return queue[(front + 1) % SIZE]
```

## 6ï¸âƒ£ í í•¨ìˆ˜êµ¬í˜„ (ì›í˜•í)

![circle](https://github.com/ykkim77/python_practice_7th/blob/main/images/circle.jpg?raw=1)

```python
#ğŸ“ ì›í˜•í ë‹¤ ì°¼ëŠ”ì§€ í™•ì¸í•˜ëŠ” í•¨ìˆ˜
def isQueueFull():
    global SIZE, queue, front, rear

    if ((rear + 1) % SIZE == front): # rearê°€ ë§ˆì§€ë§‰ì¹¸ì— ìˆëŠ”ìƒíƒœ
        return True
    else:
        return False
```

```python
#ğŸ“ ì›í˜•í ë°ì´í„°ë„£ëŠ” enQueue í•¨ìˆ˜
def enQueue(data):
    global SIZE, queue, front, rear
    
    if (isQueueFull()):
        print("íê°€ ê½‰ ì°¼ìŠµë‹ˆë‹¤.")
        return
    
    rear = (rear + 1) % SIZE # rear í•œì¹¸ ì˜®ê¸´ í›„
    queue[rear] = data
```

```python
#ğŸ“ ì›í˜•í ë°ì´í„°ë¹¼ëŠ” deQueue í•¨ìˆ˜
def deQueue():
    global SIZE, queue, front, rear

    if (isQueueEmpty()):
        print("íê°€ ë¹„ì—ˆìŠµë‹ˆë‹¤.")
        return None
    
    front = (front + 1) % SIZE
    data = queue[front]

    queue[front] = None
    return data
```
